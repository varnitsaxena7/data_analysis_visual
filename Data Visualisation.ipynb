Confirmed Covid Cases distribution of the top 10 countries...

```python
k=dataset.sort_values(by="Confirmed",ascending=False)
i=k.iloc[:10]
p=i["Confirmed"]
q=i["Country/Region"]
plt.pie(p, labels=q)
```

Hence according to this, most confirmed covid cases are from United States.

```python
mean1=dataset["Confirmed"].sum()
mean2=p.sum()
(mean2/mean1)*100
```

Hence, we can conclude that the top 10 countries of the world holds 68% of total covid cases with US, Brazil and India at the Top.   
Let us find out there individual distribution.

```python
y=k.iloc[:3]
w=y["Country/Region"]
r=y["Confirmed"]
t=r.sum()
plt.pie(r, labels=w)
```

Here United States has more than combined confirmed cases of India and Brazil.

```python
(t/mean1)*100
```

Oh! the top 3 countries of the world contributes to about 50% of the total confirmed cases of the world. It is really a very serious issue for the government of that states.

```python
ee=k.iloc[:1]
r=ee["Confirmed"]
T=r.sum()
(T/mean1)*100
```

Alas! US single handedly contributes to 26% of the total confirmed cases of the World. Hence we can now conclude that the United States is the most affected State due to Covid-19 in the world.

```python
ww=k.iloc[:5]
P=ww["Deaths / 100 Cases"]
Q=ww["Country/Region"]
plt.bar(Q,P)
plt.show()
```

Apart from the most number of Confirmed Covid Cases by US. Here Brazil has the most death rate among the top countries with most number of confirmed Covid cases. This means that Brazil has nearly the most death rate in the world.

Different Covid Rates all over the World.

```python
recover=dataset["Recovered"] 
confirm=dataset["Confirmed"] 
deaths=dataset["Deaths"]
active=dataset["Active"]
act=active.sum()
resum=recover.sum()
consum=confirm.sum()
dsum=deaths.sum()
(dsum/consum)*100
```

Death Rate Due to Covid-19 all over world=3.97%

```python
(resum/consum)*100
```

Current Recovery Rate=57.45%

```python
(act/consum)*100
```

Current Active cases rate=38.58%

```python

ans1=["Recovery Rate","Death Rate","Active Cases Rate"]
ans2=[(resum/consum)*100,(dsum/consum)*100,(act/consum)*100]
plt.title("Covid Cases Rates")
plt.bar(ans1,ans2)
```

Hence,from this we can easily conclude that of the all confirmed cases of Covid-19 that are identified all over the world 57.45% people have recovered,3.96% have died and 38.58% are currently affected due to this.

Covid Cases Conditions all over the World in Number.

```python
print("Total confirmed Cases = ",consum)
print("Total active Cases =",act)
print("Total Recovered Cases = ",resum)
print("Total Deaths = ",dsum)
```
